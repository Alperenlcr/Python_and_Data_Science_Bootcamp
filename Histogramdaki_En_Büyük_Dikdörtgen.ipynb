# Python Task - Histogramdaki En Büyük Dikdörtgen 

Her bir çubuğun genişliğinin 1 olduğu histogramın çubuk yüksekliğini temsil eden n negatif olmayan tamsayı verildiğinde, histogramdaki en büyük dikdörtgenin alanını bulun.

Yukarıda, her bir çubuğun genişliğinin 1 olduğu bir histogram verilmiştir.

height = [2,1,5,6,2,3]


image.png

Alanı 10 birim olan en büyük dikdörtgen gölgeli alanda gösterilmektedir.


**Örnek**


**Girdi:** [2,1,5,6,2,3]


**Çıktı:** 10

```python
def find_max_weight(arr):
  count = 0
  max = 0
  for i in range(len(arr)):
    if arr[i] > 0:
      count += 1
    else:
      count = 0
    if max < count:
      max = count
  return max
def largestRectangleArea(heights):
        #kodunuzu buraya yazınız
  max = 0
  count = 0
  weight = find_max_weight(heights)
  while weight != 0:
    count += 1
    if max < count*weight:
      max = count*weight
    for i in range(len(heights)):
      heights[i] -= 1
    weight = find_max_weight(heights)
  return max


if __name__ == "__main__":
    print(largestRectangleArea([2,6,4,1,2,3,7]))
    print(largestRectangleArea([2,1,5,6,2,3]))
```

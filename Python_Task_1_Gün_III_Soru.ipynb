# Python Task - 3

Bir string ifade verildiğinde, ifadenin en uzun palindromik alt dizisini bulan bir fonksiyon yazmanız gerekmektedir. Bir palindrom, ifadenin normal ve tersten okunuşunun aynı olmasıdır. 

**Not: Herhangi bir built-in fonksiyon kullanılmamalıdır.**


**Örnek:**

**Girdi:**

string = "babatenetotoses"

**Çıktı:**

En Uzun Palindrom: tenet

Uzunluk: 5


```python
def longestPalSubstr(string):
  #kodu buraya yazınız
  size = len(string)
  index = 1
  max_size = 0
  max_index = 0
  # odd solution
  for index in range(0, size-(max_size//2)):
    left = index-1
    right = index+1
    temp_size = 1
    while left >= 0 and right < size and string[left] == string[right]:
      temp_size += 2
      left -= 1
      right += 1
    if temp_size > max_size:
      max_size = temp_size
      max_index = left+1
  # even solution
  for index in range(0, size-(max_size//2)):
    left = index
    right = index+1
    temp_size = 0
    while left >= 0 and right < size and string[left] == string[right]:
      temp_size += 2
      left -= 1
      right += 1
    if temp_size > max_size:
      max_size = temp_size
      max_index = left+1
  return str(max_size)+" Palindrom : "+string[max_index : max_index + max_size]
    
     


if __name__ == '__main__':
  string = "babatenetoasdadsatosesqqqqqqqq"
  print("Length is: " + str(longestPalSubstr(string)))
```

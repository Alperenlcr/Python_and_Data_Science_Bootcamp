#Python Task - Tic Tac Toe

Tic Tac Toe (XOX) oyununu aşağıdaki yönergelere göre Class yazmanız istenmektedir.

Bir oyunda iki oyuncu olacak. Her oyuncuyu iki işaret (X ve O) temsil eder. 
Oynanış aşağıdaki gibi olmalıdır. 

•	İlk olarak, bir kullanıcı işaretini mevcut boş kutulardan birine yerleştirecektir. 

•	Ardından, ikinci kullanıcı işaretini mevcut boş kutulardan birine yerleştirecektir. 

•	Oyuncuların amacı, kendi işaretlerini tamamen sıra veya sütun şeklinde veya çapraz olarak yerleştirmektir. 

•	Oyun, bir oyuncu oyunu kazanana kadar devam eder veya kazanan bir maç olmadan tüm kutuları doldurarak berabere biter.

```python
import random


class TicTacToe:
      #kodunuzu buraya yazınız
      def __init__(self):
        self.table = [['-', '-', '-'], ['-', '-', '-'], ['-', '-', '-']]
        self.players = ['O', 'X']
        self.player = self.players[random.randint(0,1)]

      def show_table(self):
        for row in self.table:
          print(*row)

      def update_table(self, point):
        self.table[point[0]][point[1]] = self.player

      def check_end(self):
        for i in range(3):
          if (self.table[i][0] == self.table[i][1] and self.table[i][0] == self.table[i][2]) and self.table[i][2] != '-':
            return True
          if (self.table[0][i] == self.table[1][i] and self.table[0][i] == self.table[2][i]) and self.table[i][2] != '-':
            return True
        if ((self.table[0][0] == self.table[1][1] and self.table[2][2] == self.table[1][1]) or (self.table[2][0] == self.table[1][1] and self.table[0][2] == self.table[1][1])) and self.table[i][2] != '-':
          return True
        return False

      def gameplay(self):
        self.turn()
        print("Oyuncu ", self.player)
        self.show_table()
        self.update_table(self.take_input())

      def take_input(self):
        print("Satir ve Sutun numaralarini giriniz: ")
        try:
          r = list(map(int, input().split()))
        except ValueError:
          print("Hatali secim! Tekrar seciniz.")
          return self.take_input()
        if len(r) != 2 or r[0] not in [1, 2, 3] or r[1] not in [1, 2, 3]:
          print("Hatali secim! Tekrar seciniz.")
          return self.take_input()
        r[0]-=1
        r[1]-=1
        if self.table[r[0]][r[1]] != '-':
          print("Secilen yer dolu!! Tekrar seciniz.")
          return self.take_input()
        return r

      def turn(self):
        if self.player == 'O':
          self.player = 'X'
        else:
          self.player = 'O'

      def start(self):
        while not self.check_end():
          self.gameplay()
        print(self.player, "Oyuncu kazandi !!!")
        self.show_table()

if __name__ == "__main__":
  tic_tac_toe = TicTacToe()
  tic_tac_toe.start()
```

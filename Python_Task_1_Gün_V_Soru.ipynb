Satranç tahtasında verilen noktalar için bir at için olası tüm hamleleri bulan ve liste olarak döndüren bir fonksiyon yazmanız beklenmektedir.
Satranç’ta atın yapabileceği tüm hamleler aşağıdaki görsel gösterilmiştir.

**Örnek:** 

**Girdi:**

-# Atın satranç tahtası üzerindeki konumu

knight_moves((4, 4))

**Çıktı:**

-# Belirli bir konumdan at için olası tüm hareketlerin bulunduğu liste

[(6, 5), (5, 6), (2, 5), (3, 6), (2, 3), (3, 2), (5, 2), (6, 3)]


```python
def knight_moves(position):
    #kodunuzu buraya yazınız
    r = []
    for p in [(-1, -2), (-1, 2), (1, -2), (1, 2)]:
      if position[0] - p[0] >= 0 and position[0] - p[0] <= 7 and position[1] - p[1] >= 0 and position[1] - p[1] <= 7:
        r.append((position[0] - p[0], position[1] - p[1]))
      if position[0] - p[1] >= 0 and position[0] - p[1] <= 7 and position[1] - p[0] >= 0 and position[1] - p[0] <= 7:
        r.append((position[0] - p[1], position[1] - p[0]))
    return r
if __name__ == '__main__':
  
  print(knight_moves((4,4)))
```
